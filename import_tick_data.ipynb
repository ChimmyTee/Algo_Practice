```python
from datetime import datetime
import MetaTrader5 as mt5
import pandas as pd
```

```python
mt5.initialize()
```

```python

# Compute now date
from_date = datetime.now()

# Extract n Ticks before now
mt5.copy_ticks_from("EURUSD", from_date, 10, mt5.COPY_TICKS_ALL)
```

```python
# Compute now date
from_date = datetime.now()

# Extract n Ticks before now
ticks = mt5.copy_ticks_from("EURUSD", from_date, 10, mt5.COPY_TICKS_ALL)

# Transform a tuple into a dataframe - something we can understand the ticks better.
df_ticks = pd.DataFrame(ticks)

# Convert number format of the date into date format.
df_ticks["time"] = pd.to_datetime(df_ticks["time"], unit="s")

df_ticks
```

```python
# Compute now date
from_date = datetime(2023, 1, 5)
to_date = datetime(2023, 1, 6)

# Extract n Ticks before now (note that the last parameter can be modified to be something more useful
# depending on the strategy)
ticks = mt5.copy_ticks_range("EURUSD", from_date, to_date, mt5.COPY_TICKS_ALL)

# Transform Tuple into a DataFrame
df_ticks = pd.DataFrame(ticks)

# Convert number format of the date into date format.
df_ticks["time"] = pd.to_datetime(df_ticks["time"], unit="s")

df_ticks
```

```python
# Compute now date
from_date = datetime(2023, 2, 26)

# Extract rates aka candlesticks, arguments = pairs, timeframe, start, how many candles.

rates = mt5.copy_rates_from("EURUSD", mt5.TIMEFRAME_D1, from_date, 15)

# Transform Tuple into a DataFrame
df_rates = pd.DataFrame(rates)

# Convert number format of the date into date format.
df_rates["time"] = pd.to_datetime(df_rates["time"], unit="s")

df_rates
```

```python
# Compute now date
from_date = datetime(2023, 2, 20)
to_date = datetime(2023, 2, 24)

# Extract rates aka candlesticks, arguments = pairs, timeframe, start, end date.

rates = mt5.copy_rates_range("EURUSD", mt5.TIMEFRAME_H4, from_date, to_date)

# Transform Tuple into a DataFrame
df_rates = pd.DataFrame(rates)

# Convert number format of the date into date format.
df_rates["time"] = pd.to_datetime(df_rates["time"], unit="s")

df_rates
```
